```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x = [1,2,3,4,5,6,7]
y = [20,40,30,60,30,50,90]

df = pd.DataFrame({"Days": x, "No of ppl" : y})
df.head()
```

Here x is the no of days on the week 
Here y is no of ppl coming to the store on specifi days

```python
sns.lineplot(x = "Days", y = "No of ppl",data=df)
plt.show()
```

So by seeing this line plot we can understand that on 7th day more people aare coming to the store

```python
df=sns.load_dataset("tips")
df.head()
```

Here we can see that there are more datasets in the seaborn github repo in that we have choosen tips

```python
sns.lineplot(x= "day", y = "total_bill",data =df)
```

Here we have taken the data set and i am taking the days as x and totalbills as y with that data 
I have plotted the line plot 

Here if we want the plot to be more clourful and more nice we add the hue that means we add the hue as sex which male and females so  that we will get the male and female lines seperately

```python
sns.lineplot(x= "day", y = "total_bill",data =df, hue = "sex")
plt.show()
```

```python
sns.lineplot(x= "day", y = "total_bill",data =df,hue= "sex" ,style = "time")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x="who", y="fare",data = df)
```

```python
sns.barplot(x="who", y="fare",data = df, hue ="who")
```

```python
from numpy import mean
sns.barplot(x= "fare",y="who",data = df,estimator = mean)
```

```python
from numpy import median
sns.barplot(x= "fare",y="who",data = df,estimator = median)
```
