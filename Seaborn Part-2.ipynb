##HEAT MAP

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(low = 1,high =100,size=(10,10))
```

```python
print(data)
```

```python
sns.heatmap(data = data)
```

```python
sns.heatmap(data = data,linecolor = "yellow", linewidth = "2")
```

```python
sns.heatmap(data = data, annot = True)
```

```python
sns.heatmap(data = data, cbar = False)
```

```python
sns.heatmap(data = data, xticklabels= False, yticklabels=False,cbar = False)
```

#Cat plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
sns.catplot(x="sex",y="tip",data=df)
```

```python
sns.catplot(x="sex",y="tip",data=df,kind = 'bar')
```

```python
ds= sns.load_dataset("titanic")
sns.catplot(x="class",data=ds,col_wrap = 3,col ="deck")
```
